```python
df["Marca1"]      = df.Marca.apply(lambda x: 1 if x==1 else 0)
df["Marca2_y"]    = df.Marca.apply(lambda x: 1 if x==2 else 0)
df["Marca3"] =      df.Marca.apply(lambda x: 1 if x==3 else 0)
df["Marca_Inno1"] = df.Marca.apply(lambda x: 1 if x==4 else 0)
df["Marca_Inno2"] = df.Marca.apply(lambda x: 1 if x==5 else 0)

df.sample(5)
```

```python
df1 = pd.merge(dfc, df, on="Cliente")
print(df1.Cliente.nunique())
df1.sample(5)
```

```python
df1['Gerencia2'] = df1['Gerencia2'].astype(str).apply(lambda x: x.strip())
print(df1.Gerencia2.value_counts())
G2_mapping = ({"Gerencia_10": 10,
                "Gerencia_8": 8,
                "Gerencia_6": 6,
                "Gerencia_9": 9,
                "Gerencia_3": 3,
                "Gerencia_1": 1,
                "Gerencia_4": 4,
                "Gerencia_2": 2,
                "Gerencia_7": 7,
                "Gerencia_11": 11,
                "Gerencia_5": 5,
                "nan": 0})
df1["G2Map"] = df1.Gerencia2.map(G2_mapping)
print(df1.G2Map.value_counts())
```

```python
df1.sample(5)
```

```python
df1['SubCanal2'] = df1['SubCanal2'].astype(str).apply(lambda x: x.strip())
print(df1.SubCanal2.value_counts())
SC2_mapping = ({"Subcanal_3":3, "Subcanal_1":1,
                "Subcanal_4":4, "Subcanal_5":5,
                "Subcanal_8":8, "Subcanal_2":2,
                "Subcanal_11":11, "Subcanal_10":10,
                "Subcanal_6":6, "Subcanal_7":7,
                "Subcanal_9":9, "Subcanal_13":13,
                "Subcanal_31":31, "Subcanal_16":16,
                "Subcanal_14":14, "Subcanal_19":19,
                "Subcanal_12":12, "Subcanal_17":17,
                "Subcanal_18":18, "Subcanal_21":21,
                "Subcanal_20":20, "Subcanal_23":23,
                "Subcanal_26":26, "Subcanal_28":28,
                "Subcanal_22":22, "Subcanal_15":15})
df1["SC2Map"] = df1.SubCanal2.map(SC2_mapping)
print(df1.SC2Map.value_counts())
```

```python
df1['Categoria'] = df1['Categoria'].astype(str).apply(lambda x: x.strip())
print(df1.Categoria.value_counts())
Cat_mapping = ({"Categoria_2": 2,"Categoria_3":3,
                "Categoria_1": 1,"Categoria_4":4,
                "Categoria_5": 5,"Categoria_6":6})
df1["CatMap"] = df1.Categoria.map(Cat_mapping)
print(df1.CatMap.value_counts())
```

```python
df1['SegmentoPrecio2'] = df1['SegmentoPrecio2'].astype(str).apply(lambda x: x.strip())
print(df1.SegmentoPrecio2.value_counts())
Seg_mapping = ({"SegmentoPrecio_3":3,
                "SegmentoPrecio_1":1,
                "SegmentoPrecio_2":2})
df1["SegMap"] = df1.SegmentoPrecio2.map(Seg_mapping)
print(df1.SegMap.value_counts())
```

```python
df1.drop(["Gerencia2", "SubCanal2", "Categoria", "SegmentoPrecio2"], axis = 1, inplace =True)
df1.sample(5)
```

```python
df1.fillna(0)
```

```python
x = df1[["Año", "Mes", "Cliente", "nr", "disc", "Nevera", "G2Map", "CatMap", "SegMap", "Marca2_y", "Marca3", "Marca_Inno1", "Marca_Inno2"]]
y = df1["Marca1"]
df1.info()
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state=0)

x_train = preprocessing.normalize(x_train, norm='l2')
x_test = preprocessing.normalize(x_test, norm='l2')
y_train
```

```python
model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
pd.options.display.float_format = '{:.5f}'.format

y_predict = model.predict(x_test)
y_predict
```

```python
y_test
```

```python
model.score(x_test, y_test)
```

```python
print(accuracy_score(y_test, y_predict))
```

```python
recall_score(y_test, y_predict)
```

```python
precision_score(y_test, y_predict)
```

```python
f1_score(y_test, y_predict)
```

```python
y_proba = model.predict_proba(x_test)
y_proba
```
